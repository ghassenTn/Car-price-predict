```python
df.V_age.value_counts().plot(kind='bar')
plt.title('counts vehicle by age')
plt.xlabel('Age')
plt.ylabel('counts')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='V_age', hue='Fuel_Type', data=df)
plt.title('Count of Vehicle Ages by Fuel Type')
plt.xlabel('Age of Vehicle')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df.groupby(['V_age', 'Fuel_Type']).size().unstack().plot(kind='bar', stacked=True)
plt.title('Count of Vehicle Ages by Fuel Type')
plt.xlabel('Age of Vehicle')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fuel_Type', y='V_age', data=df)
plt.title('Vehicle Age Distribution by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Age of Vehicle')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Fuel_Type', y='V_age', data=df)
plt.title('Vehicle Age Distribution by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Age of Vehicle')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
df['Fuel_Type'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Vehicle Fuel Types')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='V_age', y='Fuel_Type', data=df, hue='Fuel_Type', palette='viridis', s=100)
plt.title('Scatter Plot of Vehicle Ages and Fuel Types')
plt.xlabel('Age of Vehicle')
plt.ylabel('Fuel Type')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
heatmap_data = df.groupby(['V_age', 'Fuel_Type']).size().unstack().fillna(0)
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt='g')
plt.title('Heatmap of Vehicle Ages by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Age of Vehicle')
plt.show()
```

Relation between Selling price and present price

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Present_Price', y='Selling_Price', data=df)
plt.title('Scatter Plot of Selling Price vs. Present Price')
plt.xlabel('Present Price')
plt.ylabel('Selling Price')
plt.show()
```

```python
plt.figure(figsize=(16, 6))
df.plot(x='Present_Price', y='Selling_Price', marker='o', linestyle='-', color='b')
plt.title('Line Plot of Selling Price vs. Present Price')
plt.xlabel('Present Price')
plt.ylabel('Selling Price')
plt.show()
```

```python
sns.jointplot(x='Present_Price', y='Selling_Price', data=df, kind='scatter')
plt.suptitle('Joint Plot of Selling Price vs. Present Price')
plt.show()
```

```python
sns.jointplot(x='Present_Price', y='Selling_Price', data=df, kind='hex', color='b')
plt.suptitle('Hexbin Plot of Selling Price vs. Present Price')
plt.show()
```

```python
df['Price_increase']=df['Present_Price'] - df['Selling_Price']
```

```python
df.Price_increase.max()
```

```python
df[df['Price_increase']==df.Price_increase.max()]
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Selling_Price', 'Present_Price', 'Price_increase']])
plt.title('Box Plot of Prices and Price Increase')
plt.ylabel('Price')
plt.show()
```

```python
sns.pairplot(df, height=3, aspect=1.2)
plt.suptitle('Pair Plot of Prices and Price Increase')
plt.show()
```

```python
data = {
    'Selling_Price': df.Selling_Price,
    'Present_Price': df.Present_Price,
    'Price_increase': df.Price_increase
}
df1 = pd.DataFrame(data)
sns.pairplot(df1, height=3, aspect=1.2)
plt.suptitle('Pair Plot of Prices and Price Increase')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='V_age', y='Price_increase', data=df, hue='V_age', palette='viridis', s=100)
plt.title('Scatter Plot of Price Increase vs. Age of Vehicle')
plt.xlabel('Age of Vehicle')
plt.ylabel('Price Increase')
plt.legend(title='Age')
plt.show()
```
